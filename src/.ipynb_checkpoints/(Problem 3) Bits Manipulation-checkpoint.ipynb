# Bits Manipulation

```python
# Importing relevant packages and doing necessary setup
import cv2
import matplotlib.pyplot as plt
import matplotlib
import PIL
import numpy as np

%matplotlib inline

# for auto-reloading external modules
%load_ext autoreload
%autoreload 2
```

```python
# Helper function to display multiple images in a plot
def display_images_normalized(images, rows, cols, size, title, subtitles):
    fig = plt.figure(figsize = size)
    c = 0
    sz = len(images)
    for i in range(rows):
        for j in range(cols):
            if sz == c:
                break
            ax1 = fig.add_subplot(rows, cols, c + 1)
            ax1.imshow(images[c], cmap = 'gray')
            ax1.set_title(subtitles[c], fontsize = 20)
            c = c + 1
    plt.show()
```

### Part 1

```python
# Precompute powers of 2
P = []
P.append(1)
for i in range(1, 10):
    P.append(P[i - 1] * 2)
```

```python
def BitQuantizeImage(im, k):
    img = np.zeros(im.shape)
    bit = k - 1
    for i in range(8, 8 - k, -1):
        if isinstance(im[0][0], np.ndarray) == False:
            # Processing a grayscale image
            img = img + ((((1 << (i - 1)) & im) > 0) * P[bit])
        else:
            # Processing a color image
            # Change in index is due to normalization!
            img = img + ((((1 << (i - 1)) & im) > 0) * P[i - 1])
        bit = bit - 1
    img = img.astype(np.uint8)
    return img
```

```python
img = cv2.cvtColor(cv2.imread('../input_data/quantize.jpg'), cv2.COLOR_BGR2RGB)
images = []
images.append(img)
images.append(BitQuantizeImage(img, 1))
images.append(BitQuantizeImage(img, 2))
images.append(BitQuantizeImage(img, 3))
images.append(BitQuantizeImage(img, 4))
images.append(BitQuantizeImage(img, 5))
images.append(BitQuantizeImage(img, 6))
images.append(BitQuantizeImage(img, 7))
images.append(BitQuantizeImage(img, 8))
display_images_normalized(images, 3, 3, (20, 20), 'title', ['Original Image', '1 bit', '2 bits', '3 bits', '4 bits', '5 bits', '6 bits', '7 bits', '8 bits'])
```

### Part 2

```python
def BitPlaneImage(im, k):
    """ 0 based indexing for k i.e planes are numbered from 0-7 """
    img = np.zeros(im.shape)
    img = (((1 << k) & im) > 0) * 1
    return img.astype(np.uint8)
```

```python
img = cv2.imread('../input_data/cameraman.png', cv2.IMREAD_GRAYSCALE)
images = []
images.append(img)
images.append(BitPlaneImage(img, 0))
images.append(BitPlaneImage(img, 1))
images.append(BitPlaneImage(img, 2))
images.append(BitPlaneImage(img, 3))
images.append(BitPlaneImage(img, 4))
images.append(BitPlaneImage(img, 5))
images.append(BitPlaneImage(img, 6))
images.append(BitPlaneImage(img, 7))
display_images_normalized(images, 3, 3, (20, 20), 'title', ['Original Image', '0th bit plane', '1st bit plane', '2nd bit plane', '3rd bit plane', '4th bit plane', '5th bit plane', '6th bit plane', '7th bit plane'])
```

### Part 3

For lena1.jpg, the 4th bit plane(0 based indexing) of the original image of Lena is displayed as verified below

```python
img = cv2.imread('../input_data/lena.jpg', cv2.IMREAD_GRAYSCALE)
checker = cv2.imread('../input_data/lena1.jpg', cv2.IMREAD_GRAYSCALE)
output_img = BitPlaneImage(img, 4)
images = []
images.append(img)
images.append(output_img)
images.append(checker)
display_images_normalized(images, 2, 2, (15, 15), 'title', ['Original Image of Lena', '4th bit plane of Original image', 'Ground Truth'])
```

For lena2.jpg, the original image of Lena is quantised using 2 bits as verified below

```python
img = cv2.imread('../input_data/lena.jpg', cv2.IMREAD_GRAYSCALE)
checker = cv2.imread('../input_data/lena2.jpg', cv2.IMREAD_GRAYSCALE)
output_img = BitQuantizeImage(img, 2)
images = []
images.append(img)
images.append(output_img)
images.append(checker)
display_images_normalized(images, 2, 2, (15, 15), 'title', ['Original Image of Lena', 'Original image quantised with 2 bits', 'Ground Truth'])
```
