# Histogram Equalization and Matching

```python
# Importing relevant packages and doing necessary setup
import cv2
import matplotlib.pyplot as plt
import matplotlib
import PIL
import numpy as np

%matplotlib inline

# for auto-reloading external modules
%load_ext autoreload
%autoreload 2
```

```python
# Helper function to display multiple images in a plot
def display_images_unnormalized(images, rows, cols, size, title, subtitles):
    fig = plt.figure(figsize = size)
    c = 0
    sz = len(images)
    for i in range(rows):
        for j in range(cols):
            if sz == c:
                break
            ax1 = fig.add_subplot(rows, cols, c + 1)
            ax1.imshow(images[c], cmap = 'gray', vmin = 0, vmax = 255)
            ax1.set_title(subtitles[c], fontsize = 20)
            c = c + 1
    plt.show()
```

### Part 1

```python
def histEqualization(im, L):
    sz = (im.shape[0] * im.shape[1])
    cum_prob = np.zeros(256)
    for i in range(256):
        cum_prob[i] = sum(sum((im <= i))) * 1.0 / sz
    img = np.round((L - 1) * cum_prob[im])
    return img.astype(np.uint8)
```

### Part 2

```python
path = ['../input_data/hist_equalization1.png', 
'../input_data/hist_equalization2.png', '../input_data/hist_equalization3.png',
'../input_data/hist_equalization4.png']
images = []
subtitles = []
for i in range(len(path)):
    img = cv2.imread(path[i], cv2.IMREAD_GRAYSCALE)
    images.append(img)
    subtitles.append('Original Image' + str(i + 1))
    subtitles.append('Histogram Equalised Image' + str(i + 1))
    images.append(histEqualization(img, 256))
display_images_unnormalized(images, len(path), 2, (10, 24), 'title', subtitles)
```

Histogram equalisation tries to assign the same number of pixels for intensities in an image. As a result the number of low intensity and high intensity pixels are balanced and images appear enhanced.<br> 
For original images 1, 2 and 3; their intensity distribution is concentrated only in the darker side. After applying histogram equalization we spread the same distribution across a wider range of intensities.<br>
We can see clearly that histogram equalization doesn't have much effect when images already have a well spread equal intensity distribution over all intensities as evident from Image 4.

### Part 3

```python
def histMatching(im, ref_im):
    img_rs = histEqualization(im, 256)
    sz = (ref_im.shape[0] * ref_im.shape[1])
    cum_prob = np.zeros(256)
    for i in range(256):
        cum_prob[i] = sum(sum((ref_im <= i))) * 1.0 / sz
    img_rs = img_rs / 255
    res = np.searchsorted(cum_prob, img_rs)
    return res.astype(np.uint8)
```

```python
img = cv2.imread('../input_data/eye.png', cv2.IMREAD_GRAYSCALE)
ref_im = cv2.imread('../input_data/eyeref.png', cv2.IMREAD_GRAYSCALE)
output = histMatching(img, ref_im)
images = []
subtitles = []
images.append(img)
images.append(ref_im)
images.append(output)
subtitles.append('Original Image')
subtitles.append('Reference Image')
subtitles.append('Histogram Matched Image')
display_images_unnormalized(images, 2, 2, (10, 10), 'title', subtitles)
```

### Part 4

```python
c = 1
path = ['../input_data/part1.png', 
'../input_data/part2.png', '../input_data/part3.png',
'../input_data/part4.png']
images = []
for i in range(len(path)):
    img = cv2.imread(path[i], cv2.IMREAD_GRAYSCALE)
    images.append(histEqualization(img, 256))
# Match the histogram of images[1] with that of images[0]
for i in range(3):
    images[i] = histMatching(images[i], images[3])
output = np.zeros([len(images[0]) + len(images[2]), len(images[0][0]) + len(images[1][0])])
output[0: len(images[0]), 0:len(images[0][0])] = images[0]
output[0: len(images[1]), len(images[0][0]):] = images[1]
output[len(images[0]): , 0:len(images[2][0])] = images[2]
output[len(images[1]): , len(images[2][0]):] = images[3]
output = output.astype(np.uint8)
output = histEqualization(output, 256)
images = []
subtitles = ['']
images.append(output)
display_images_unnormalized(images, 1, 1, (10, 24), 'title', subtitles)
```
