# Contrast Stretching

```python
# Importing relevant packages and doing necessary setup
import cv2
import matplotlib.pyplot as plt
import matplotlib
import PIL
import numpy as np

%matplotlib inline

# for auto-reloading external modules
%load_ext autoreload
%autoreload 2
```

```python
# Helper function to display multiple grayscale images in a plot
def display_grayscale_images_unnormalized(images, rows, cols, size, title, subtitles):
    fig = plt.figure(figsize = size)
    c = 0
    sz = len(images)
    for i in range(rows):
        for j in range(cols):
            if sz == c:
                break
            ax1 = fig.add_subplot(rows, cols, c + 1)
            ax1.imshow(images[c], cmap = 'gray', vmin = 0, vmax = 255)
            ax1.set_title(subtitles[c], fontsize = 20)
            c = c + 1
    plt.show()
```

### Part 1

```python
def linContrastStretching(im, a, b):
    mn = np.amin(im)
    mx = np.amax(im)
    im = im.astype(np.float64)
    img = (im - mn) * (b - a) * 1.0 / (mx - mn)
    img = np.round(img + a, decimals=0)
    return img.astype(np.uint8)
```

### Part 2

```python
img = cv2.imread('../input_data/cameraman.png', cv2.IMREAD_GRAYSCALE)
output_img = linContrastStretching(img, 200, 255)
images = []
images.append(img)
images.append(output_img)
display_grayscale_images_unnormalized(images, 1, 2, (10, 20), 'x', ['Original Image', 'Intensity: [200, 255]'])
```
