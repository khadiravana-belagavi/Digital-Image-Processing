```python
# Helper function to display multiple images in a plot
def display_images(images, rows, cols, size, title, subtitles):
    fig = plt.figure(figsize = size)
    c = 0
    sz = len(images)
    for i in range(rows):
        for j in range(cols):
            if sz == c:
                break
            ax1 = fig.add_subplot(rows, cols, c + 1)
            ax1.imshow(images[c], cmap = 'gray')
            ax1.set_title(subtitles[c], fontsize = 20)
            c = c + 1
    plt.show()
    # Un comment the below line and comment the above line to save and display
    #plt.savefig('part2.png')
```

```python
colors = {}
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if tuple(img[i][j]) in colors.keys():
            colors[tuple(img[i][j])] += 1
        else:
            colors[tuple(img[i][j])] = 1
colors = sorted(colors.items(), key = 
             lambda kv:(kv[1], kv[0]))
colors = colors[len(colors) - 6 : len(colors) - 1]
print(colors)
```

## Part 2

```python
img = cv2.imread('../input_data/5_2_1.jpeg', cv2.IMREAD_GRAYSCALE)
img = (img < 90).astype(np.uint8)
# Appropriately cropping
img = img[30: img.shape[0] - 30, 40: img.shape[1] - 100]
visited = np.zeros(img.shape)
images = []
subtitles = []
images.append(img)
subtitles.append('Original Image')
display_images(images, 1, 1, (20, 20), '', subtitles)
```

```python
def dfs(row, col, color):
    if row < 0 or col < 0 or row >= img.shape[0] or col >= img.shape[1]:
        return 0
    if img[row][col] == 0:
        return 0
    if visited[row][col] != 0:
        return 0
    visited[row][col] = color
    return (1 + dfs(row + 1, col, color) + dfs(row - 1, col, color) + dfs(row, col + 1, color) + dfs(row, col - 1, color) + dfs(row + 1, col + 1, color) + dfs(row + 1, col - 1, color) + dfs(row - 1, col + 1, color) + dfs(row - 1, col - 1, color))
```

```python
comp = 1
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if visited[i][j] == 0 and img[i][j] == 1:
            dfs(i, j, comp)
            comp += 1
print(comp)
```

```python
def hole_frequency(t):
    output = np.zeros(img.shape)
    x_min = 1000000000000
    x_max = -10000000000000
    y_min = 1000000000000
    y_max = -10000000000000
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            if visited[i][j] == t:
                output[i][j] = 1
                x_min = min(x_min, i)
                x_max = max(x_max, i)
                y_min = min(y_min, j)
                y_max = max(y_max, j)
    c = 1
    output = output[x_min - 5: x_max + 5, y_min - 5: y_max + 5]
    visited2 = np.zeros(output.shape)
    output = 1 - output
    def dfs2(row, col, color):
        if row < 0 or col < 0 or row >= output.shape[0] or col >= output.shape[1]:
            return 0
        if output[row][col] == 0:
            return 0
        if visited2[row][col] != 0:
            return 0
        visited2[row][col] = color
        return (1 + dfs2(row + 1, col, color) + dfs2(row - 1, col, color) + dfs2(row, col + 1, color) + dfs2(row, col - 1, color) + dfs2(row + 1, col + 1, color) + dfs2(row + 1, col - 1, color) + dfs2(row - 1, col + 1, color) + dfs2(row - 1, col - 1, color))
    for i in range(output.shape[0]):
        for j in range(output.shape[1]):
            if visited2[i][j] == 0 and output[i][j] == 1:
                dfs2(i, j, c)
                c += 1
    c = c - 2
    return c
```

```python
OUTPUT = np.zeros(img.shape)
for i in range(1, comp):
    if hole_frequency(i) == 1:
        for j in range(img.shape[0]):
            for k in range(img.shape[1]):
                if visited[j][k] == i:
                    OUTPUT[j][k] = 1
images = []
subtitles = []
images.append(OUTPUT)
subtitles.append('Letters that enclose one empty region')
display_images(images, 1, 1, (20, 20), '', subtitles)
```

```python
OUTPUT = np.zeros(img.shape)
for i in range(1, comp):
    if hole_frequency(i) == 2:
        for j in range(img.shape[0]):
            for k in range(img.shape[1]):
                if visited[j][k] == i:
                    OUTPUT[j][k] = 1
images = []
subtitles = []
images.append(OUTPUT)
subtitles.append('Letters that enclose two empty regions')
display_images(images, 1, 1, (20, 20), '', subtitles)
```
