```python
# Helper function to display multiple images in a plot
def display_images(images, rows, cols, size, title, subtitles):
    fig = plt.figure(figsize = size)
    c = 0
    sz = len(images)
    for i in range(rows):
        for j in range(cols):
            if sz == c:
                break
            ax1 = fig.add_subplot(rows, cols, c + 1)
            ax1.imshow(images[c], cmap = 'gray')
            ax1.set_title(subtitles[c], fontsize = 20)
            c = c + 1
    plt.show()
    # Un comment the below line and comment the above line to save and display
    #plt.savefig('part2.png')
```

### Part 1

```python
def RECURSIVEonedFFT(a):
    n = a.shape[0]
    if n == 1:
        y = a
        return y
    a_even = a[0 : n : 2]
    a_odd = a[1 : n : 2]
    y_even = RECURSIVEonedFFT(a_even);
    y_odd = RECURSIVEonedFFT(a_odd);
    E = []
    for k in range(int(n / 2)):
        E.append(np.exp(np.complex(0, -1) * 2.0 * np.pi * k / n))
    E = np.array(E)
    y = []
    y = y_even + E * y_odd
    y = np.append(y, y_even - E * y_odd)
    return y
```

```python
arr = np.array([1,2,3,4])
print(RECURSIVEonedFFT(arr))
print(np.fft.fft(arr))
```

### Part 2

```python
def RECURSIVEFFT1D(a):
    n = a.shape[1]
    if n == 1:
        y = a
        return y
    a_even = a[:, 0 : n : 2]
    a_odd = a[:, 1 : n : 2]
    y_even = RECURSIVEFFT1D(a_even);
    y_odd = RECURSIVEFFT1D(a_odd);
    E = []
    for k in range(int(n / 2)):
        E.append(np.exp(np.complex(0, -1) * 2.0 * np.pi * k / n))
    E = np.array(E)
    y = []
    y = y_even + E * y_odd
    y = np.append(y, y_even - E * y_odd, axis = 1)
    return y
```

```python
def RECURSIVEtwodFFT(I):
    colffts = RECURSIVEFFT1D(I)
    colffts = np.transpose(colffts)
    output = RECURSIVEFFT1D(colffts)
    output = np.transpose(output)
    return output
```

```python
arr = np.array([[1,2], [3,4]])
print(RECURSIVEtwodFFT(arr))
print(np.fft.fft2(arr))
```

```python
img = cv2.imread('../input_data/lena.jpg', cv2.IMREAD_GRAYSCALE)
images = []
subtitles = []
images.append(img)
images.append(np.log(1 + np.abs(np.fft.fftshift(np.fft.fft2(img)))))
images.append(np.log(1 + np.abs(np.fft.fftshift(RECURSIVEtwodFFT(img)))))
subtitles.append('Original Image(lena.jpg)')
subtitles.append('Frequency Spectrum using inbuilt function')
subtitles.append('Frequency Spectrum using my own two dimensional FFT function')
display_images(images, 2, 2, (20, 20), '', subtitles)
```
