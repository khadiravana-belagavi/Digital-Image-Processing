# Contrast Stretching

```python
# Importing relevant packages and doing necessary setup
import cv2
import matplotlib.pyplot as plt
import matplotlib
import PIL
import numpy as np

%matplotlib inline

# for auto-reloading external modules
%load_ext autoreload
%autoreload 2
```

```python
# Helper function to display multiple grayscale images in a plot
def display_grayscale_images_unnormalized(images, rows, cols, size, title, subtitles):
    fig = plt.figure(figsize = size)
    c = 0
    sz = len(images)
    for i in range(rows):
        for j in range(cols):
            if sz == c:
                break
            ax1 = fig.add_subplot(rows, cols, c + 1)
            ax1.imshow(images[c], cmap = 'gray', vmin = 0, vmax = 255)
            ax1.set_title(subtitles[c], fontsize = 20)
            c = c + 1
    plt.show()
```

### Part 1

```python
def linContrastStretching(im, a, b):
    mn = np.amin(im)
    mx = np.amax(im)
    im = im.astype(np.float64)
    img = (im - mn) * (b - a) * 1.0 / (mx - mn)
    img = np.round(img + a, decimals=0)
    return img.astype(np.uint8)
```

### Part 2

```python
# This helper function is required to generate the colorbars required for this part
def colorbar(im, k):
    u, c = np.unique(im, return_counts=True)
    t = list(zip(c, u))
    t = sorted(t, key = lambda x : x[0])
    lst = []
    idx = len(t) - 1
    for i in range(min(len(t), k)):
        lst.append(t[idx][1])
        idx = idx - 1
    width = 10
    res = np.zeros((width, len(lst) * width))
    c = 0
    for i in lst:
        for j in range(c, c + width):
            for k in range(width):
                res[k][j] = i
        c = c + width
    return res.astype(np.uint8)
```

```python
img = cv2.imread('../input_data/lena_low_contrast.jpeg', cv2.IMREAD_GRAYSCALE)
output_img = linContrastStretching(img, 0, 255)
images = []
images.append(img)
images.append(output_img)
images.append(colorbar(img, 15))
images.append(colorbar(output_img, 15))
display_grayscale_images_unnormalized(images, 2, 2, (10, 10), 'x', ['Original Image', 'Intensity: [0, 255]', 'Colorbar for above image', 'Colorbar for above image'])
```

#### Explanation for the resulting colorbars above
We can see clearly that the colorbar for the contrast stretched image is more prominent than the colorbar for the original image. We are almost unable to distinguish between the colors of the first three squares of the colorbar for the original image where as there is a clear distinction between the colors of the first three squares of the colorbar of the contrast stretched image. Similar observations can be made for some other squares also. We have clearly enhanced the distinction of similar colors i.e enhanced the contrast. We can also explain why the dark colors become darker and the light colors become lighter by the argument that the contrast is stretched over a larger range of intensities 

### Part 3

```python
img = cv2.imread('../input_data/cameraman.png', cv2.IMREAD_GRAYSCALE)
output_img = linContrastStretching(img, 0, 255)
images = []
subtitles = []
images.append(img)
images.append(output_img)
subtitles.append('Original Image 1')
subtitles.append('Contrast Stretched Image 1')
img = cv2.imread('../input_data/hist_equalization3.png', cv2.IMREAD_GRAYSCALE)
output_img = linContrastStretching(img, 0, 255)
images.append(img)
images.append(output_img)
subtitles.append('Original Image 2')
subtitles.append('Contrast Stretched Image 2')

img = cv2.imread('../input_data/c1.jpg', cv2.IMREAD_GRAYSCALE)
output_img = linContrastStretching(img, 0, 255)
images.append(img)
images.append(output_img)
subtitles.append('Original Image 3')
subtitles.append('Contrast Stretched Image 3')

img = cv2.imread('../input_data/c2.jpg', cv2.IMREAD_GRAYSCALE)
output_img = linContrastStretching(img, 0, 255)
images.append(img)
images.append(output_img)
subtitles.append('Original Image 4')
subtitles.append('Contrast Stretched Image 4')

img = cv2.imread('../input_data/cameraman.png', cv2.IMREAD_GRAYSCALE)
output_img = linContrastStretching(img, 200, 255)
images.append(img)
images.append(output_img)
subtitles.append('Original Image 5')
subtitles.append('Contrast Stretched Image 5')
display_grayscale_images_unnormalized(images, 5, 2, (30, 30), 'x', subtitles)
```

The effect is not so much on image 1 because the image already has well balanced number of pixels with intensities covering the entire range of 0 to 255.<br>
The effect is not so much on image 2 although majority of the pixels are on the darker side i.e low intensity. If we carefully observe there are very very few intensities on the higher side of the spectrum as well. This is not evident because the number of pixels with high intensities is very small compared to number of pixels with low intensities and therefore we tend to miss them; however they are there. Since contrast enhancement uses maximum and minimum intensity to do the transformation even if one pixel is there with an intensity of 255 and if there is another pixel with intensity of 0, the image will be considered as fully stretched.<br>
Images 3 and 4 are best examples where the effect is prominent. Image 3 has high intensity pixels and image 4 has low intensity pixels. Their intensity ranges are stretched to cover the entire spectrum of intensities from 0 to 255<br>
Contrast Stretching has a prominent effect on image 5 though it has a well balanced number of pixels with intensities covering the entire spectrum. This is because the contrast is stretched to a smaller range of intensities and hence majority of pixels undergo a transformation.
