```python
# Helper function to display multiple images in a plot
def display_images(images, rows, cols, size, title, subtitles):
    fig = plt.figure(figsize = size)
    c = 0
    sz = len(images)
    for i in range(rows):
        for j in range(cols):
            if sz == c:
                break
            ax1 = fig.add_subplot(rows, cols, c + 1)
            ax1.imshow(images[c], cmap = 'gray')
            ax1.set_title(subtitles[c], fontsize = 20)
            c = c + 1
    plt.show()
    # Un comment the below line and comment the above line to save and display
    #plt.savefig('part2.png')
```

```python
def convolve(im, k):
    im = im.astype(np.float64)
    k = k.astype(np.float64)
    rows = im.shape[0]
    cols = im.shape[1]
    k_r = k.shape[0]
    k_c = k.shape[1]
    res = np.zeros(im.shape)
    for i in range(int(k_r/2), rows - int((k_r-1) / 2)):
        for j in range(int(k_c/2), cols - int((k_c - 1) / 2)):
            l_r = i - int(k_r / 2)
            r_r = i + int((k_r - 1) / 2)
            l_c = j - int(k_c / 2)
            r_c = j + int((k_c - 1) / 2)
            res[i][j] = max(0, sum(sum(k * im[l_r : r_r + 1, l_c : r_c + 1])))
    return res.astype(np.int)
```

### Part 1

```python
img = cv2.imread('../input_data/box.png', cv2.IMREAD_GRAYSCALE)
kernel = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])
output = convolve(img, kernel)
images = []
subtitles = []
images.append(img)
images.append(output)
subtitles.append('Original Image')
subtitles.append('Convolved Image')
display_images(images, 1, 2, (10, 10), 'title', subtitles)
```

```python
img = cv2.imread('../input_data/blur.jpg', cv2.IMREAD_GRAYSCALE)
kernel = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])
output = convolve(img, kernel)
output2 = convolve(img, kernel.transpose())
images = []
subtitles = []
images.append(img)
images.append(output)
images.append(output2)
subtitles.append('Original Image')
subtitles.append('Convolved with the above matrix')
subtitles.append('Convolved with transpose of above matrix')
display_images(images, 3, 1, (30, 30), 'title', subtitles)
```

Convolving the image with the matrix $$
\left(\begin{array}{cc} 
-1 & -1 & -1\\
0 & 0 & 0\\
1 & 1 & 1
\end{array}\right)
$$
(which we had designed for part 1) is equivalent to finding the y-gradient. Convolving the image with the transpose of the above matrix is equivalent to finding the x-gradient.
<br>
The Y Gradient captures changes in intensity along the Y Axis. The X Gradient captures changes in intensity along the X Axis.
